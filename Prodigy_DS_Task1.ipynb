```python
## Plotting a histogram to visualize the distribution of passenger ages

import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/train.csv')
plt.hist(df['Age'],color='Red',edgecolor='Black',alpha=0.9,bins=10,histtype='bar')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequecncy')
plt.show()
```

```python
# Creating a bar chart to show the frequency of each gender in the dataset

df['Sex'].value_counts().plot(color='Skyblue',kind='bar', edgecolor='black'
,alpha=0.8)
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.show()
```
